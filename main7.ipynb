```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as st
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('data_7_1.csv',sep=';', encoding='cp1251', na_values=['NA', 'NULL', ' '])
# df = pd.read_csv('data.csv', sep=',', na_values=['NA', 'NULL', ' '])
df.columns = df.columns.str.lower().str.replace(' ', '_')
#df = df.dropna()  # удалили строки с пропусками
df.head(30)
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.drop('продажи', axis=1), df['продажи'], test_size=0.15, random_state = 42)
lin_regression = LinearRegression()
lin_regression.fit(x_train, y_train)
prediction_test = lin_regression.predict(x_test)
print(prediction_test.round())
```

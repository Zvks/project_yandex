```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

lst = ['Покупательская активность', 'Тип сервиса', 'Разрешить сообщать', 'Маркет_актив_6_мес', 'Длительность', 'Акционные_покупки', 'Популярная_категория', 'Средний_просмотр_категорий_за_визит', 'Выручка_от_клиента_текущий_месяц', 'Разность_выручки_тек_прошлый_месяц', 'Ошибка_сервиса', 'Страниц_за_визит', 'Всего_минут_текущем_месяце', 'Всего_минут_предыдущий_месяц']
for i in lst:
    name = i.lower().replace(' ', '_')
    df = pd.read_csv('upload.csv', sep=',', na_values=['NA', 'NULL', ' '], encoding = '1251')
    df.columns = df.columns.str.lower().str.replace(' ', '_')
    y = df[name].value_counts().sort_index()
    plt.pie(y, labels=y.index, autopct='%1.1f%%')
    plt.show()
    #plt.savefig('target_2_5.png')
```

```python
import pandas as pd
import numpy as np
from scipy import stats

def eta_squared(x, x):
  results = stats.f_oneway(*[y[x == cat] for cat in x.unique()])
  ssb = sum([len(y[x == cat]) * (np.mean(y[x == cat]) - np.mean(y))**2 for cat in x.unique()])
  sst = sum((y - np.mean(y))**2)

  eta2 = ssb / sst
  return eta2


    
df = pd.read_csv('data_2.csv', sep=',', na_values=['NA', 'NULL', ' '], encoding = '1251')
df.columns = df.columns.str.lower().str.replace(' ', '_')
df = df.dropna()
print(eta_squared(df["ингредиент_1"], df["брак"]))
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


def create_scatter_plot(df, y_factor):
    sns.scatterplot(x='вес', y=y_factor, hue='оператор_линии', data=df)

    plt.title('Диаграмма рассеяния:\nАнализ зависимости между факторами.',
              fontsize=16,  # Размер шрифта заголовка
              y=1.01)  # Расстояние от заголовка до графика

    plt.xlabel('Вес')
    plt.ylabel(y_factor)
    plt.legend(title='Оператор линии') # Optional: Add title to legend

    plt.show()


y_factor = input()
df = pd.read_csv('data_2.csv', sep=',', na_values=['NA', 'NULL', ' '], encoding = '1251')
df.columns = df.columns.str.lower().str.replace(' ', '_')
df = df.dropna()
create_scatter_plot(df, y_factor)
```
